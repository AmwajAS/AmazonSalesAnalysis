4. Data Cleaning

```python
#Changing Data Types of Columns from object to ___
#columns : discounted_price, actual_price, discount_percentage
#discounted_price
#cleaning the column discounted_price from the sign appeared and the comma.
df['discounted_price'] = df['discounted_price'].str.replace("₹", '')
df['discounted_price'] = df['discounted_price'].str.replace(",", '')
df['discounted_price'] = df['discounted_price'].astype('float64')
```

```python
#actual_price
df['actual_price'] = df['actual_price'].str.replace("₹", '')
df['actual_price'] = df['actual_price'].str.replace(",", '')
df['actual_price'] = df['actual_price'].astype('float64')
```

```python
#discount_percentage
df['discount_percentage'] = df['discount_percentage'].str.replace("%", '')   #sign removed in order to convert the column type
df['discount_percentage']= df['discount_percentage'].astype('float64')
df['discount_percentage'] = df['discount_percentage'] /100        #mathematical operations or analysis, percentages are usually represented as decimals
                                                    
```

```python
#Handle the rating column:
# Get the frequency of each unique rating value
rating_frequency = df['rating'].value_counts()

# Display the frequency of each rating value
print(rating_frequency)

#get the record:
df.query('rating == "|"')
```

```python
#replacing the value
df['rating'] = df['rating'].str.replace('|', '3.9')
df.query('rating =="3.9"')
```

```python
#Convert dataType for rating:
df['rating'] = df['rating'].astype('float64')
```

```python
#rating_count

df['rating_count'] = df['rating_count'].str.replace(",",'').astype('float')
```

Converting Categorical labels into Numeric values that allows models to capture patterns and relationships between these categorical features and the target variable.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df.info()
```

```python
#label encode for the Categorical values
le_product_id = LabelEncoder()
le_product_name = LabelEncoder()
le_category = LabelEncoder()
le_about_product = LabelEncoder()
le_user_id = LabelEncoder()
le_user_name = LabelEncoder()
le_review_id = LabelEncoder()
le_review_title = LabelEncoder()
le_review_content = LabelEncoder()
le_product_name = LabelEncoder()
le_user_name = LabelEncoder()
le_user_id = LabelEncoder()
le_img_link = LabelEncoder()
le_product_link = LabelEncoder()

#fit the LabelEncoder on the dataset and then transform the categories into their corresponding numeric values.

df['product_id'] = le_product_id.fit_transform(df['product_id'])
df['category'] = le_category.fit_transform(df['category'])
df['review_id'] = le_review_id.fit_transform(df['review_id'])
df['review_content'] = le_review_content.fit_transform(df['review_content'])
df['product_name'] = le_product_name.fit_transform(df['product_name'])
df['user_name'] = le_user_name.fit_transform(df['user_name'])
df['about_product'] = le_about_product.fit_transform(df['about_product'])
df['user_id'] = le_user_id.fit_transform(df['user_id'])
df['review_title'] = le_review_title.fit_transform(df['review_title'])
df['img_link'] = le_img_link.fit_transform(df['img_link'])
df['product_link'] = le_product_link.fit_transform(df['product_link'])
```

```python
#correlation heatmaps

plt.figure(figsize=(25,25))
plt.xticks(fontsize=14, rotation=45) 
plt.yticks(fontsize=14)
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot =True)
plt.show()
```

Key Points:
Correlation Coefficient:
- Range: The correlation values range from -1 to 1.
   +1: A perfect positive correlation. When one variable increases, the other increases proportionally.
   0: No correlation. The variables are not related to each other.
   -1: A perfect negative correlation. When one variable increases, the other decreases proportionally.
- Significance:
  Strong positive or negative values (closer to +1 or -1) indicate a strong linear relationship.
  Values closer to 0 indicate weaker relationships.

heatmap Observations:
- The strongest correlation is between the actual_price & discounted_price - expected
-  negative correlation between discounted_price and discount_percentage - As the discount_percentage higher the lower the  discounted_price gets.
-  No significane Correlation Between rating and rating_count
-  Weak Negative Correlation of -0.15: between rating and discount_percentage -  items with higher discounts may not have the highest ratings, which   could be explained by the fact that products that don't perform as well might be given higher discounts to increase sales

```python
#further analysis of the relationships:

#  Pearson correlation coefficients 
correlation_matrix = df.corr()
print(correlation_matrix)

plt.figure(figsize=(25,25))
plt.xticks(fontsize=14, rotation=45) 
plt.yticks(fontsize=14)
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix (Pearson)")
plt.show()

#  Spearman correlation coefficients (for non-linear relationships)
spearman_correlation_matrix = df.corr(method="spearman")

# Print the Spearman correlation matrix
print(spearman_correlation_matrix)

# Create a heatmap to visualize the Spearman correlations
plt.figure(figsize=(25,25))
plt.xticks(fontsize=14, rotation=45) 
plt.yticks(fontsize=14)
sns.heatmap(spearman_correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix (Spearman)")
plt.show()
```

4. What are the Top 5 categories based with highest ratings?


```python
average_ratings = df.groupby("category")["rating"].mean().reset_index()

# Sort by average rating in descending order
average_ratings = average_ratings.sort_values(by="rating", ascending=False)

# Print the top 5 categories
print("Top 5 categories with highest average ratings:")
for i in range(5):
    category = average_ratings.iloc[i]["category"]
    average_rating = average_ratings.iloc[i]["rating"]
    print(f"{i+1}. {category}: {average_rating:.2f}")
```
